```python
def sorter(dataframe):
    
    a = dataframe.iloc[:6:]
    b = dataframe.iloc[6:18:]
    c = dataframe.iloc[18:30:]
    d = dataframe.iloc[30:42:]
    e = dataframe.iloc[42:54:]
    f = dataframe.iloc[54:66:]
    g = dataframe.iloc[66:78:]
    
    returns = [a,b,c,d,e,f,g]
    
    raw_list = []
    for x in returns:

        mean_blah = x.mean()
        raw_list.append(mean_blah)
        
    std_list = []
    for x in returns:
    
        std_var = x.std()
        std_list.append(std_var)
    
    print(raw_list)
    print(std_list)
```
